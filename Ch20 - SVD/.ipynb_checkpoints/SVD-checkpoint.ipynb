```python
import numpy as np

m, n = 2, 3
A = np.random.rand(m, n)
U, S, V = np.linalg.svd(A)

# Check if U, V are orthogonal and S is a diagonal matrix with nonnegative decreasing value
print('Frobenius norm of (UU^T - I):', np.linalg.norm(U.dot(U.T) - np.eye(m)))
print('S =', S)
print('Frobenius norm of (VV^T - I):', np.linalg.norm(V.dot(V.T) - np.eye(n)))
```

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

img = mpimg.imread('D:/HUST/AI/MLCB/Code/14 - Singular Value Decomposition/building2.jpg')
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# to gray 
gray = 0.2125* img[:, :, 0] + 0.7154 *img[:, :, 1] + 0.0721 *img[:, :, 2]
plt.imshow(gray)
plt.axis('off')
plt.show()
```

```python
U, S, V = np.linalg.svd(gray)

from matplotlib.backends.backend_pdf import PdfPages

with PdfPages('singular_value.pdf') as pdf:
    plt.semilogy(S) 
    plt.xlabel('$k$')
    plt.ylabel('$\sigma_k$')
    plt.show()
```

```python
# percentage of preserving energy

with PdfPages('energy_preserved.pdf') as pdf:
    a = np.sum(S**2)
    b = np.zeros_like(S)
    for i in range(S.shape[0]):
        b[i] = np.sum(S[:i+1]**2, axis = 0)/a

    plt.plot(b)
    plt.xlabel('$k$', fontsize = 20)
    plt.ylabel('$\|\mathbf{A} - \mathbf{A}_k\|_F^2$  /  $\|\mathbf{A}\|_F^2$', fontsize = 20)
    plt.show()
```

```python
def approx_rank_k(U, S, V, k):
    Uk = U[:, :k]
    Sk = S[:k]
    Vk = V[:k, :]
    return np.around(Uk.dot(np.diag(Sk)).dot(Vk))

e = 1 - b
img1 = approx_rank_k(U, S, V, 5)
plt.imshow(img1)
plt.axis('off')
plt.title('k = 5, error = ' + str(e[0]))
plt.show()

img2 = approx_rank_k(U, S, V, 50)
plt.imshow(img2)
plt.axis('off')
plt.title('k = 50, error = ' + str(e[9]))
plt.show()

img3 = approx_rank_k(U, S, V, 100)
plt.imshow(img3)
plt.axis('off')
plt.title('k = 100, error = ' + str(e[19]))
plt.show()
```
