# Setup

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

# Linear Regression Example

```python
np.random.seed(4)
N_train = 30
N_test = 20

X_train = np.random.rand(N_train, 1)*5 # np.random.rand: random samples from a uniform distribution over [0, 1)
y_train = 3*(X_train - 2)*(X_train - 3)*(X_train - 4) + 10*np.random.randn(N_train, 1)
# np.random.randn: random samples from a Gaussian distribution of mean 0 and variance 1

X_test = (np.random.rand(N_test, 1) - 1/8)*10
y_test = 3*(X_test - 2)*(X_test - 3)*(X_test - 4) + 10 *np.random.randn(N_test, 1)

def buildXBar(X, d): # Create feature vector XBar = (1, x, x^2, ..., x^d)
    X_Bar = np.ones((X.shape[0], 1))
    for i in range(1, d + 1):
        X_Bar = np.concatenate((X_Bar, X**i), axis=1)
    return X_Bar


def myFit(d):
    plt.plot(X_train, y_train, 'ro', label='Training samples')
    plt.plot(X_test, y_test, 'yo', label='Test samples')
    
    x = np.linspace(-2, 7, 100, endpoint=True)
    y = 3*(x - 2)*(x - 3)*(x - 4)
    plt.plot(x, y, 'g', label='True model')
    
    X_train_Bar = buildXBar(X_train, d)
    regr = linear_model.LinearRegression(fit_intercept=False)
    regr.fit(X_train_Bar, y_train)
    w = regr.coef_

    y_trained_model = np.zeros_like(x)
    for i in range(d + 1):
        y_trained_model += w[0][i]*x**i

    plt.plot(x, y_trained_model, 'b', label='Trained model')    
    
    plt.xticks([], [])
    plt.yticks([], [])
    plt.axis([-4, 10, np.amin(y_test) - 100, np.amax(y_test) + 100])
    plt.title('d=' + str(d))
    plt.legend()
    plt.show()
 
myFit(2)
myFit(4)
myFit(8)
myFit(16)
```

# Validation Example

```python
np.random.seed(5)
N = 50
N_test = 10

X = np.random.rand(N, 1)*5
y = 3*(X - 2)*(X - 3)*(X - 4) + 10*np.random.randn(N, 1)
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.33, random_state=0)

X_test = np.random.rand(N_test, 1)*5
y_test = 3*(X_test - 2)*(X_test - 3)*(X_test - 4) + 10*np.random.randn(N_test, 1)

def buildXBar(X, d):
    XBar = np.ones((X.shape[0], 1))
    for i in range(1, d + 1):
        XBar= np.concatenate((XBar, X**i), axis=1)
    return XBar

# Calculating y = f(X) = a_0 + a_1*x + a_2*x^2 + ... + a_n*x^n
#                      = a_0 + x(a_1 + x(a_2 + ... +x(a_n-1 + a_n*x)))
def poly(a, X): 
    p = np.zeros_like(X)
    for i in range(len(a) - 1, -1, -1):
        p = p*X + a[i]
    return p

def MSE(X, y, w):
    y_pred = poly(w, X)
    return np.mean(np.abs(y - y_pred))

def myFit(d):
    X_train_Bar = buildXBar(X_train, d)
    regr = linear_model.LinearRegression(fit_intercept=False)
    regr.fit(X_train_Bar, y_train)
    w = regr.coef_[0]
    
    train_err = MSE(X_train, y_train, w)
    valid_err = MSE(X_valid, y_valid, w)
    test_err = MSE(X_test, y_test, w)
    
    return train_err, valid_err, test_err

d = 9
train_error = []
valid_error = []
test_error = []
for i in range(1, d):
    (train_err, valid_err, test_err) = myFit(i)
    train_error.append(train_err)
    valid_error.append(valid_err)
    test_error.append(test_err)
    
degree = range(1, d)
plt.plot(degree, train_error, 'b-', label='Training error')
plt.plot(degree, valid_error, 'r-', label='Validation error')
plt.plot(degree, test_error, 'g-', label='Test error')

plt.xlabel('Degree')
plt.ylabel('Error')
plt.legend()
plt.show()
    
```
